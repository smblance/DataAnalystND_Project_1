## 0. The data.
The data consists of the performance in the Stroop test for 24 participants.<br />
Each row shows result in seconds in the congruent and incongruent tasks for one participant.

```python
import pandas as pd
import numpy as np
import scipy.stats
from ggplot import *
%matplotlib inline
data = pd.read_csv('stroopdata.csv')
print data
```

## 1. The independent and dependent variables.
Independent variable: Congruency (whether the color of the word matches the word). <br />
Dependent variable: Performance (measured by the time it takes to name the colors of all the words).

## 2. Set of hypotheses for the task.
To test whether congruency matters, we will test whether the conguent and incongruent samples come from the same distribution.<br />
To do this, we will first test the two samples for normality. We will do it with the scipy.stats' "Normaltest", which is based on the analysis of skew and kurtosis of the distribution, and also with the Shapiro-Wilk test.<br />
If both of the samples will pass the normality test, we will use the Student's T-test, and if either sample does not pass - the Mann-Whitney U-test.<br />
So, the null hypothesis is: "The congruent and incongruent samples come from the same distribution."<br />
The alternative hypothesis: "The congruent and incongruent samples come from different distributions."

## 3. Descriptive statistics regarding this dataset.

```python
print data.describe()
print '\nCongruent median:',data.Congruent.median()
print 'Incongruent median:', data.Incongruent.median()
print '\nCongruent range: ', (data.max() - data.min()).Congruent
print 'Incongruent range: ', (data.max() - data.min()).Incongruent
```

We can see that means, medians, ranges and standard deviations are very different between the congruent and incongruent tasks.

## 4. Visualizations that show the distribution of the sample data.

```python
print ggplot(aes(), data = data) \
    + geom_histogram(aes('Congruent'), alpha = 0.5, fill = 'red', binwidth = 0.6) \
    + geom_histogram(aes('Incongruent'), alpha = 0.5, fill = 'blue', binwidth = 0.6) \
    + xlab('Performance in seconds\n\nRed: Congruent, Blue: Incongruent') \
    + ylab('Frequency') \
    + ggtitle ('Histogram of congruent and incongruent performance')
```

Here we can see that the incongruent sample is clearly shifted to the right, comparing to the congruent.

```python
print ggplot(aes(0), data = pd.DataFrame(data.Incongruent - data.Congruent)) \
    + geom_histogram(fill = 'purple', alpha = 0.5, binwidth = 0.5) \
    + xlab('Difference in performance, seconds') \
    + ylab('Frequency') \
    + ggtitle ('Histogram of Difference in congruent and incongruent performance of subjects')
```

From this plot we can conclude that in most cases, the difference between the two results is somewhere between 3 and 12 seconds.<br />
We also see that there is no participant who performed in the incongruent scenario better than in the congruent.<br />
To infer this relation onto the population, we need to confirm it with a proper test.<br />
We also see two outliers - participants #14 and #19, who have the difference in time between tasks of about 17 and 22 seconds, respectively.<br />
These two participants did not show particularly good performance in the congruent task, and it is plausible that they performed much worse in the incongruent one.<br />
2 outliers out of 24 (8%) is a reasonable amount and can be explained by variation in the populaton.

```python
slope, intercept, r_value, _, _ = scipy.stats.linregress(data.Congruent, data.Incongruent)
print "Coefficient of determination: ", r_value**2
print ggplot(data, aes()) \
    + geom_point(aes(x = 'Congruent',y = 'Incongruent'), color = 'red', size = (data.Incongruent - data.Congruent)*5, alpha = 0.7) \
    + geom_abline(intercept = 0, slope = 1, color = 'blue') \
    + geom_abline(intercept = intercept, slope = slope, color = 'red') \
    + xlab('Congruent performance \n\nSize illustrates difference in performance \n Line illustrates where the points should be if Congruency does not matter') \
    + ylab('Incongruent performance') \
    + ggtitle ('Difference in Congruent and Incongruent performance of subjects')
```

On this scatterplot the blue line shows the region where the points would lie if the congruency does not matter.<br /> 
The red line is the linear regression of the points, with R-squared value of 0.12.<br />
The size of the point indicates the (vertical) distance from the blue line.<br />
The two outliers can be clearly seen.

## 5. Statistical tests and results.

```python
from scipy import stats
print 'Congruent normality: \n', stats.normaltest(data.Congruent), '\nShapirotestResult', stats.shapiro(data.Congruent)
print '\nIncongruent normality: \n', stats.normaltest(data.Incongruent), '\nShapirotestResult',stats.shapiro(data.Incongruent)
#print '\nVariance equality: \n', stats.levene(data.Congruent, data.Incongruent), '\n', stats.bartlett(data.Congruent, data.Incongruent)
```

We see that both normality tests produce the same results: Congruent data is normal, and Incongruent is not.<br />
So, instead of the Student's T-test we will use (one-sided) Mann-Whitney U-test.

```python
print 'Mann-Whitney U-test:\n', stats.mannwhitneyu(data.Congruent,data.Incongruent)
```

This lets us reject the null hypothesis (U = 37.0, p < 0.000001) and conclude that the Incongruent test results are significantly higher than the Congruent, which matches our expectations.

## 6. What is responsible for the effects observed? What alternative or similar tasks would result in similar effect?

The main theory behind the effect is that the two streams of information, the color of the text and the meaning of the word, are processed in parallel, and the meaning is processed faster. When the meaning is processed, the brain is ready for response selection, and it takes time and effort to:
- understand that the response is wrong, 
- wait for the color to be processed,
- finally choose the correct response.<br />

We are used to a lot of reading, so word processing is done much more automatically and needs less attention than color processing.<br />
If we change the task, asking the subject to report the information that is processed faster (word meaning), than the lag before the answer is much smaller, because the response selection can be done before the conflicting information (color) is even processed.

Some of the stimuli that we can use to design similar tests are:
- color of the word
- meaning of word (can be colors/numbers/shapes/etc.)
- emotion of the word
- shape
- size of the word or individual letters
- sound of the word or sound of some animal or natural phenomenon, e.g. rain
- number of the words or objects, e.g."55555"
- location of the object inside a box

So, the following tests are expected to produce effects similar to the Stroop test:
- name the length of the given string of numbers. Correct answers will be: "555" -> 3, "22222" -> 5
- name the location of the word inside the square. If the word "left" is at the bottom side of the square, the subject must report "bottom".
- name the number with the biggest size. If "2" is written in the larger font than "5", the subject must report "2".
